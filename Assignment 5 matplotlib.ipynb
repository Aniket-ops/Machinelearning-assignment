```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df=pd.read_csv('survey_results_public.csv')
df.head(3)
```

```python
i = df[df['Country'] == 'India']
i
```

```python
i.dropna(inplace=True)
i
```

```python
plt.title('India Devs Age')
plt.xlabel('Age')
plt.ylabel('Total')
arr = np.arange(10, 90, 10)
plt.hist(i['Age'], bins=arr, edgecolor='black')
```
